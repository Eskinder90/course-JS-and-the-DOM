```python
import math
import matplotlib.pyplot as plt

# Given values
r = 0.1275 # radius of the scaled globe in meters
interval_x = 20 # interval for meridians (longitude) in degrees
interval_y = 10 # interval for parallels (latitude) in degrees

def calculate_parallel_spacing(r, theta):
    """Calculate spacing of parallels (latitude lines) from the equator."""
    theta_radians = math.radians(theta) # Convert degrees to radians
    return r * math.sin(theta_radians)

# Generate latitudes and longitudes
latitudes = range(-90, 91, interval_y) # Latitude range from -90 to 90 degrees
longitudes = range(-180, 181, interval_x) # Longitude range from -180 to 180 degrees

# Compute Y (latitude) spacings
latitude_lines = [calculate_parallel_spacing(r, lat) for lat in latitudes]

# Compute X (longitude) spacings (equal spacing for meridians)
longitude_spacing = [r * math.radians(lon) for lon in longitudes]

# Plotting the Lambert Equal-Area Projection
fig, ax = plt.subplots(figsize=(12, 8))

# Plot latitude lines (horizontal lines)
for lat, y in zip(latitudes, latitude_lines):
    ax.hlines(y=y, xmin=-r * math.pi, xmax=r * math.pi, colors='blue', linestyles='--', label=f"Latitude {lat}°" if lat == 0 else None)

# Plot longitude lines (vertical lines)
for lon, x in zip(longitudes, longitude_spacing):
    ax.vlines(x=x, ymin=latitude_lines[0], ymax=latitude_lines[-1], colors='red', linestyles='-', label=f"Longitude {lon}°" if lon == 0 else None)

# Add labels and titles
ax.set_title("Lambert Equal-Area Projection Grid", fontsize=16)
ax.set_xlabel("East-West Spacing (meters)", fontsize=12)
ax.set_ylabel("North-South Spacing (meters)", fontsize=12)
ax.legend(loc='upper left', fontsize=10)

# Grid and axis adjustments
ax.grid(True, linestyle='--', alpha=0.7)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import math

# Parameters
r = 0.1275  # Scaled globe radius in meters
interval_x = 20  # Interval for meridians (longitude) in degrees
interval_y = 10  # Interval for parallels (latitude) in degrees

# Function to calculate parallel spacing
def calculate_parallel_spacing(r, theta):
    theta_radians = math.radians(theta)
    return r * math.sin(theta_radians)

# Generate latitude and longitude values
latitudes = range(-90, 91, interval_y)  # Latitudes from -90° to 90°
longitudes = range(-180, 181, interval_x)  # Longitudes from -180° to 180°

# Compute parallel and meridian spacings
latitude_positions = [calculate_parallel_spacing(r, lat) for lat in latitudes]
longitude_positions = [r * math.radians(lon) for lon in longitudes]

# Display the results
print("Latitude Spacings (y-coordinates):")
for lat, y in zip(latitudes, latitude_positions):
    print(f"Latitude {lat}° -> y = {y:.4f} meters")

print("\nLongitude Spacings (x-coordinates):")
for lon, x in zip(longitudes, longitude_positions):
    print(f"Longitude {lon}° -> x = {x:.4f} meters")
# Plotting the Lambert Equal-Area Projection
fig, ax = plt.subplots(figsize=(12, 8))

# Plot latitude lines (horizontal lines)
for lat, y in zip(latitudes, latitude_lines):
    ax.hlines(y=y, xmin=-r * math.pi, xmax=r * math.pi, colors='blue', linestyles='--', label=f"Latitude {lat}°" if lat == 0 else None)

# Plot longitude lines (vertical lines)
for lon, x in zip(longitudes, longitude_spacing):
    ax.vlines(x=x, ymin=latitude_lines[0], ymax=latitude_lines[-1], colors='red', linestyles='-', label=f"Longitude {lon}°" if lon == 0 else None)

# Add labels and titles
ax.set_title("Lambert Equal-Area Projection Grid", fontsize=16)
ax.set_xlabel("East-West Spacing (meters)", fontsize=12)
ax.set_ylabel("North-South Spacing (meters)", fontsize=12)
ax.legend(loc='upper left', fontsize=10)

# Grid and axis adjustments
ax.grid(True, linestyle='--', alpha=0.7)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
plt.tight_layout()

# Show plot
plt.show()
```

```python
# Parameters
r = 0.1275  # Scaled radius of the globe in meters
interval_theta = 10  # Interval for latitude (in degrees)
interval_phi = 20  # Interval for longitude (in degrees)

# Function to calculate parallel spacing
def calculate_gnomonic_parallel_spacing(r, theta):
    theta_radians = math.radians(theta)
    return r * math.tan(theta_radians) if abs(theta) < 90 else float('inf')

# Generate latitude and longitude values
latitudes = range(10, 90, interval_theta)  # Latitudes excluding 0° and 90°
longitudes = range(0, 360, interval_phi)  # Longitudes from 0° to 360°

# Compute parallel spacings
latitude_radii = [calculate_gnomonic_parallel_spacing(r, lat) for lat in latitudes]

# Compute longitude spacings
longitude_angles = [math.radians(lon) for lon in longitudes]

# Display results
print("Latitude Spacings (y-coordinates):")
for lat, y in zip(latitudes, latitude_radii):
    print(f"Latitude {lat}° -> y = {y:.4f} meters")

print("\nLongitude Angles (in radians for plotting):")
for lon, angle in zip(longitudes, longitude_angles):
    print(f"Longitude {lon}° -> angle = {angle:.4f} radians")

# Plotting the Gnomonic Projection
fig, ax = plt.subplots(figsize=(10, 10))

# Draw parallels as concentric circles
for radius, lat in zip(latitude_radii, latitudes):
    circle = plt.Circle((0, 0), radius, color='blue', fill=False, linestyle='--', label=f"Latitude {lat}°" if lat == 10 else None)
    ax.add_artist(circle)

# Draw meridians as radiating lines
for lon in longitudes:
    angle = math.radians(lon)
    x = latitude_radii[-1] * math.cos(angle)
    y = latitude_radii[-1] * math.sin(angle)
    ax.plot([0, x], [0, y], color='red', linestyle='-', label=f"Longitude {lon}°" if lon == 0 else None)

# Add labels and titles
ax.set_title("Normal Gnomonic Projection", fontsize=16)
ax.set_xlabel("X (meters)", fontsize=12)
ax.set_ylabel("Y (meters)", fontsize=12)
ax.legend(loc='upper left', fontsize=10)

# Set aspect ratio and limits
ax.set_aspect('equal', adjustable='datalim')
plt.xlim(-1.5 * r, 1.5 * r)
plt.ylim(-1.5 * r, 1.5 * r)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Show plot
plt.show()
```
